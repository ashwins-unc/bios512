Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
library('scales')
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res=350)

p2 = ggplot(penguins) +
    geom_histogram(fill='#3182bd', alpha=0.7, color='#3182bd') +
    aes(x=body_mass_g) +
    labs( title = "Penguin Species Mass by Year", y="Number of Species", x="Body Mass (g)" ) +
    facet_wrap(~year, ncol=1) +
    theme_minimal() +
    scale_x_continuous(label=comma) +
    guides(x = guide_axis(angle = 45)) +
    scale_y_continuous()
    
p2
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
library('palmerpenguins')
options(repr.plot.height=6, repr.plot.width=4, repr.plot.res=300)

p3 = ggplot(penguins) +
    geom_point( aes(x=body_mass_g, y=sex), position=position_jitter(h=0, w=0.3)) +
    facet_wrap(~species, ncol=1) +
    labs(title = "Body Mass for Penguins by Species", x="Sex", y="Body Mass (g)") +
    scale_y_continuous(label=comma) +
    theme_minimal()

p3
```
